```python
import numpy as np
from sklearn.datasets import load_iris,make_classification
import matplotlib.pyplot as plt
```

```python
data = load_iris()
```

```python
data
```

```python
X = data.data[data.target!=2]
X = np.insert(X, 4, 1, axis=1)
```

```python
Y = data.target[data.target!=2]
Y
```

```python
Y = Y.reshape((-1,1))
```

```python
class LogisticRegression():
    def sigmoid(self, x):
        return 1/(1+np.exp(-x))
        
    def update(self, X, y):
        temp = np.zeros((X.shape[1],1))
        for i in range(X.shape[0]):
            temp += (X[i]*(np.exp(X[i]@self.w)/(1+np.exp(X[i]@self.w))-Y[i])).reshape((-1,1))
        self.w -= 0.01*temp
        
    def fit(self, X, y):
        self.w = np.random.randn(X.shape[1], 1)
        for i in range(100):
            self.update(X, y)
            y_pred = self.sigmoid(X @ self.w)
            print(np.mean(0.5*(y_pred-y)**2))
            
            
    def predict(self, X):
        y_pred = np.round(self.sigmoid(X @ self.w))
        return y_pred  
```

参数的更新也可简写成 **w -= 0.01 * X.T @ (y_pred - y)**，和update()的效果一样

```python
classifier = LogisticRegression()
```

```python
classifier.fit(X,Y)
```

```python
y_pred = classifier.predict(X)
y_pred
```

```python
accuracy = np.sum(Y == y_pred, axis=0) / len(Y)
```
